```python
label = LabelEncoder()
```

```python
df_train.drop('Cabin',axis=1,inplace=True)
df_test.drop('Cabin',axis=1,inplace=True)
```

```python
df_train['Embarked'] = df_train['Embarked'].fillna(np.random.choice(a = ['S','C','Q'],p = [0.7244094488188977,0.1889763779527559,0.08661417322834646]))
```

```python
df_train.drop('Ticket',axis=1,inplace=True)
df_test.drop('Ticket',axis=1,inplace=True)
```

```python
df_train['Sex'] = label.fit_transform(df_train['Sex'])
df_test['Sex'] = label.transform(df_test['Sex'])
```

```python
df_train['Embarked'] = label.fit_transform(df_train['Embarked'])
df_test['Embarked'] = label.transform(df_test['Embarked'])
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import GridSearchCV
```

```python
from sklearn.metrics import classification_report,ConfusionMatrixDisplay,confusion_matrix
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
X_train = df_train.drop('Survived',axis=1)
y_train = df_train['Survived']
X_test = df_test
```

```python
scaler = StandardScaler()
model = LogisticRegression()
```

```python
from sklearn.pipeline import Pipeline
```

```python
pipe = Pipeline([('scaler',scaler),('model',model)])
```

```python
params = {
    'model__penalty' : ['l2','l1','elastinetic',None],
    'model__solver' : ['lbfgs', 'liblinear', 'newton-cg', 'newton-cholesky', 'sag', 'saga'],
    'model__max_iter' : [100, 500, 1000]
}
```

```python
final_model = GridSearchCV(pipe, param_grid=params, cv=3, verbose=2)
```

```python
final_model.fit(X_train, y_train)
```

```python
final_model.best_params_
```

```python
y_pred = final_model.predict(X_test)
```

```python
df_sub = pd.read_csv('gender_submission.csv')
```

```python
df_sub['Survived'] = y_pred
```

```python
df_sub.to_csv('sub2.csv',index=False,index_label=False)
```
